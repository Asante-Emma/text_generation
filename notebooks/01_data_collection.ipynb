```python
import pandas as pd
import glob
import zipfile
import os
```

```python
# Define file paths
zip_file_path = '../data/raw/archive (7).zip'
extract_to_path = '../data/raw'

# Unzip the files
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_to_path)

# Load csv files using glob.glob
csv_files = glob.glob(os.path.join(extract_to_path, "csv/*.csv"))

# Combine all CSV files into a single DataFrame
df_list = [pd.read_csv(file) for file in csv_files]
combined_df = pd.concat(df_list, ignore_index=True)

# Save the combined data
combined_df.to_csv('../data/processed/combined_lyrics.csv', index=False)
```
